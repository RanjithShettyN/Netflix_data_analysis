```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
netflix_data = pd.read_csv("netflix_titles.csv")


```

```python
# Display the first few rows of the dataset
print(netflix_data.head())

```

```python
# 1. Basic Information about the Data

# Check for missing values
print("\nMissing values:\n", netflix_data.isnull().sum())

```

```python
# Data type of each column
print("\nData Types:\n", netflix_data.dtypes)
```

```python

# 2. EDA & Visualizations

# Distribution of Movies vs TV Shows
plt.figure(figsize=(8,6))
sns.countplot(x="type", data=netflix_data)
plt.title("Distribution of Movies vs TV Shows")
plt.show()

```

```python
# Distribution of Content Ratings
plt.figure(figsize=(12,6))
sns.countplot(x="rating", data=netflix_data, order=netflix_data['rating'].value_counts().index)
plt.title("Distribution of Content Ratings")
plt.xticks(rotation=45)
plt.show()

```

```python
# Top 10 Countries producing content
plt.figure(figsize=(12,6))
top_countries = netflix_data['country'].value_counts().head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 10 Countries Producing Content")
plt.xticks(rotation=45)
plt.show()

```

```python
# Content added over the years
plt.figure(figsize=(12,6))
sns.countplot(x="release_year", data=netflix_data, order=netflix_data['release_year'].value_counts().index)
plt.title("Content Added Over the Years")
plt.xticks(rotation=90)
plt.show()

```

```python
# Genre distribution for Movies
plt.figure(figsize=(15,7))
list_of_genres = netflix_data[netflix_data["type"] == "Movie"]["listed_in"].str.split(',').sum()
sns.countplot(y=list_of_genres, order=pd.Series(list_of_genres).value_counts().index)
plt.title("Genre Distribution for Movies")
plt.show()

```

```python
# Genre distribution for TV Shows
plt.figure(figsize=(15,7))
list_of_genres_tv = netflix_data[netflix_data["type"] == "TV Show"]["listed_in"].str.split(',').sum()
sns.countplot(y=list_of_genres_tv, order=pd.Series(list_of_genres_tv).value_counts().index)
plt.title("Genre Distribution for TV Shows")
plt.show()

# You can continue with more analysis based on the columns and information available in the dataset.

# Conclusion:
# From the above visualizations, we get a basic understanding of the kind of content available on Netflix. 
# More detailed analysis can be done depending on specific objectives or questions one might have about the data.
```
